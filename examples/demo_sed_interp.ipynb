# Example to call the SED function

```python
import numpy as np
import matplotlib.pyplot as plt
from MorphSED.morphsed import sed_bc03

import matplotlib as mpl
mpl.rc("xtick", direction="in", labelsize=18)
mpl.rc("ytick", direction="in", labelsize=18)
mpl.rc("xtick.major", width=1., size=8)
mpl.rc("ytick.major", width=1., size=8)
mpl.rc("xtick.minor", width=1., size=5)
mpl.rc("ytick.minor", width=1., size=5)
```

```python
wave = np.linspace(2000, 5e4, 1000)
z = 0.04  # metallicity
age = [0.5, 1, 5, 9]  # Gyr
logMs = 0

lum = []
for loop, a in enumerate(age):
    lum.append(sed_bc03(wave, z, a, logMs))

fig = plt.figure(figsize=(8, 5))
ax = plt.gca()

for loop, l in enumerate(lum):
    ax.plot(wave, l, color='C{0}'.format(loop), label='{0} Gyr'.format(age[loop]))
ax.legend(loc='upper right', fontsize=18)
ax.set_yscale('log')
ax.set_ylim([5e31, 2e34])
ax.set_xlabel(r'Wavelength ($\mathrm{\AA}$)', fontsize=24)
ax.set_ylabel(r'$\nu L_\nu (\mathrm{erg\,s^{-1}}$)', fontsize=24)
ax.minorticks_on()
plt.show()
```
